- *P-1.29* Write a Python program that outputs all possibles strings formed by using the characters ` 'c', 'a', 't','d','o', and 'g' ` exactly once. 

```python
def permutations(iterable: list) -> tuple:
    """
    returns the possible permutations of a sequence
    """
   
    items = [*iterable] #Unpacking the iterable into a list.
    
    def find_permutations(start_idx: int) -> list[tuple]:
        """
        swaps the items in the list and returns the new combination.
        """
        
        if start_idx == len(items) - 1: ##check if the list is empty, 
            yield tuple(items)           # if so, return it.    
            
        for i in range(start_idx, len(items)):
            #Using recursion returns possible permutations 
            #Swapping the elements in the list.
            
            items[start_idx], items[i] = items[i], items[start_idx]
            yield from find_permutations(start_idx + 1)
            items[start_idx], items[i] = items[i], items[start_idx]
    
    yield from find_permutations(0) #starts the recursion 
    

string = 'catdog'
test = 'abc'
next_ = iter(permutations(string))

print(list(permutations(test)))
print(next(next_))
print(next(next_))
```

- *P-1.30* Write a Python program that can take a positive integer greater that 2 as input and write out the number of times one mist repeatedly divide this number by 2 before getting a value less than 2.

```python
def descom(number: int) -> int:
    
    times = len(bin(number)[2:]) - 2
    print(f"the number {number} can be divided by two {times} times before it is less than two.")
    print(f"its minimum expression before it is less than two is { number >> times}")
    
    return ''

print(descom(25), "\n")
print(descom(152), "\n")
print(descom(2345), "\n")
```

- *P-1.31* Write a Python program that can "make change." Your program should take two numbers as input, one that is a monetary amount charged and the other that is monetary amount given. It should then return the number of each kind of bill and coin to give back as change for the difference between the amount given and the amount charged. The values assigned to the bills and coins can be based on the monetary system of any current or former government. Try to design your program so that it returns as few bills and coins as possible.

```python

import keyboard

def number_decomposer(number: int | float, key= None) -> list:
    """
        n > 0 \\
        decomposition of the number into its relative values. \\
        return a list in the order: millions, thousands, hundreds, ones and decimal part.\\
        key --> place value  e.g. "decimal" or list/tuple of positional values e.g. ["tens", "ones", "millions",...]
    """

    # Go to my project "Numbers to Words" in jupyter's notebook "Dictionaries and Sets"  
    # to better understand this function.
    units = number
    tens = hundreds = thousands = millions = int()
    
    #In a single loop decomposes the number into its relative values
    while units >= 10: 
        if units >= 1_000_000:
            millions = units // 1_000_000
            units -= millions * 1_000_000
        
        if units >= 1_000:
            thousands = units // 1_000
            units -=  thousands * 1_000
        
        if units >= 100:
            hundreds = units // 100
            units -= hundreds * 100

        if units >= 10 :
            tens = units // 10
            units -=  tens * 10
    
    place_values = [ int(millions) * pow(10,6) , int(thousands ) * pow(10,3) , int(hundreds ) * pow(10,2) , 
                    int(tens ) * 10 , int(units), round(units - int(units), 2)] 
    match key :
        case list() | tuple() | str() as key:
            place_values = {"decimal": place_values[5], "ones": place_values[4] , "tens": place_values[3], 
                            "hundreds": place_values[2], "thousands": place_values[1], "millions": place_values[0]}
            
            return list(map(lambda x : place_values.get(x), key)) if isinstance(key,(list,tuple)) else place_values.get(key)
        
        case _:
            return place_values
       
def displays(money: int | float ) -> print:
    """Displays the number of banknotes and coins needed to represent the money """
    
    print(f"{'banknotes':>5}{'amount':>9}")   if money >= 1 else print(f"{'coins':>5}{'amount':>13}")

    #If the number is greater than one, it is passed to the function, otherwise it is multiplied by 100 so that the decimal part becomes an integer and thus easier to represent.
    number_to_display = number_decomposer(money)  if money >= 1 else  number_decomposer(money * 100) # This will work in the recursive call (below) if the number has a decimal part.
    

    for number in number_to_display: #[millions, thousands, hundreds, ones, decimal part]
        
        if number :
            if number >= 1_000: # Displays the number of thousand notes needed to represent the number
                print(f"{1000:>5} {number // 1_000:>9}") 
            
            elif number >= 100: #Displays the number of hundred notes needed to represent the number, 
                 print(f"{100:>5} {number // 100:>9}")
                 
            elif number >= 10: #If there are units greater than ten, then
                
                if number == 50: #If it is equal to 50, only a 50 note is needed
                     print(f"{50:>5} {number // 50:>9}")
                
                elif number < 50: # If it is less than 50, then either n-notes of ten or only one are needed
                     print(f"{10:>5} {number // 10:>9}")
                     
                else: #If it is over 50, then only one 50 note and one/s of 10 is needed
                     print(f"{50:>5} {number // 50:>9}") 
                     print(f"{10:>5} {(number - 50) // 10:9}")
                        
            elif number < 1 : #If it is less than 1, then it is its decimal part.
                print() # Therefore, the function is called recursively 
                displays(number) # to print the decimal part as if it were an integer (above)
            
            else: #If no conditions are met, this means that it is less than ten and greater than or equal to one. 
                 print(f"{1:>5} {number // 1:>9}") #Therefore, it shows the number of one's banknotes that are needed

def make_change():
    """ Shows the number of banknotes needed to represent a monetary amount \\
        obtained from the difference between the money received and the price charged.
    """
    try:
        monetary_amount = float(input("Enter monetary amount given: ")) , float(input("Enter monetary amount charged : "))
        result = round( monetary_amount[0] - monetary_amount[1], 2)

        print()
        print(f"Monetary given   \u00A3{monetary_amount[0]:^2,.2f}")
        print(f"Monetary charged \u00A3{monetary_amount[1]:^2,.2f}")
        print(f"Change equal to  \u00A3{result:^2,.2f}\n")
       
        displays(abs(result)) #The built-in abs function is used in case the amount received is less than the amount charged, that is, a negative number.
   
    except:
       print()
       print("Error")
       print("the value entered must be an integer or decimal")
       
print("Welcome")
make_change() # Stars the 'Program'

while True:
    condition = input("press enter to perform another operation or escape to exit the program. ")
    
    if keyboard.is_pressed("enter")  :
        make_change()
        
    elif keyboard.is_pressed("esc"):
        break
```

- P-1.33 Write a Python program that simulates a handheld calculator. Your program should process input from the Python console representing buttons that are "pushed," and then output the contents of the screen after each operation is performed. Minimally, your calculator should be able to process the basic arithmetic operations and a reset/clear operation

```python

from tkinter import *

root = Tk()
root.geometry("411x533")
root.configure(background='#1a1a1a') 
root.attributes('-alpha', 0.97)
root.resizable(False, False)
root.title("Calculator")

def clear_display(CE = False, undo = False) -> None:
  '''It clears the input field and superfield.
  
  Parameters
  ----------
  CE, optional
    Clear Entry
  undo, optional
    If True, the last character in the input field will be deleted.
  
  '''
  
  if CE:
    input_field.delete(0, END) #clear the entries in input field 
  
  elif undo:
    input_text.set(input_field.get()[:-1])  #erase the entries in input field
    
  else :
    #Clear the entries in input field and superfield.
    input_field.delete(0, END) 
    input_superfield.delete(0, END)
    

def get_numbers(number)  -> None:
  '''Inserts the number in input field'''

  if  '=' in input_superfield.get() : # If the equals symbol is in the SuperField, all content on the screen is deleted. 
    clear_display()
    
  elif "\t" in input_superfield.get(): # After inserting any operator, 
    input_field.delete(0, END) # the expression (i.e. the left operand and the operator) is placed in the super field,
    input_superfield.delete(0,1)# then what is in the InputField is removed so that the right operand  can be placed in the Input Field. 
    # Note: the character '\t' is used as a sentinel. This is to know when to remove the Input Field and when not to remove it.(see function get_operation)
  
  display_state = input_field.get() + str(number) 
  input_text.set(display_state)
  

def get_operation(operator) -> None:
  '''Inserts the operator in superfield'''
  
  text_in_superfield = "\t" + input_field.get() + " " + operator # The left operand and operator are obtained and then inserted in the SuperField.
  #Note:The character "is used so that after deleting the input field it is possible to insert the right operand with more than one digit.(see function get_numbers)
  
  input_superfield.delete(0, END) if  input_superfield.get()  else None # If an expression or operator already exists in the SuperField, it is removed.
  
  superscript_text.set(text_in_superfield) # The left operand and the operator are inserted in the SuperField.

    
def calculate(operator = str()) -> None:
  '''It takes the current state of the input field and the superfield, 
    and evaluates the expression and displays the result in the input field.
  
  Parameters
  ----------
  operator: optional 
    the unary operator to be used in the calculation (e.g: =, n^2, 1/n, sqrt(n), +/-, %, etc)

  '''
  import re
  
  # Initialization Phase: Creation of Expressions #
  
  if not operator :
    if '=' not in input_superfield.get():
      display_state = input_superfield.get() + " " + input_field.get()  #This creates a standard expression, for example: 2 + 4 = 6
      #This occurs when first inserting the left operand, the operator, the right operand and then pressing the equals symbol. 
   
    else :
       # This expression is created when the right operand and the same operator are reused, e.g. A) 2 + 6 = 8, B) 8 + 6 = 14, ....
       #This occurs when, after performing the first operation, the equal symbol is pressed again.
      display_state =  input_field.get() + ' '.join(re.findall(r'([-+*/]+)(\d+|\d+\.\d+)(=)', input_superfield.get().replace(' ',''))[0]) 
    
  else :
    display_state = input_field.get() + operator if operator != "1/" else operator + input_field.get() # Creates the expression when there is only one operand and an unary operator, e.g.: 4**2 , 1/16, sqrt(4)
    input_superfield.delete(0, END) if  "=" in input_superfield.get()  else None  # If an expression already exists in the SuperField, it is removed.

  # Processing Phase: Realization of the expression #
  
  result = round(eval(re.sub('[,=]+', "", display_state)), 2)  # The expression is evaluated and the result is obtained. 
  
  # Termination Phase: Displays the Result of the expression #
  
  if '=' not in input_superfield.get() and not operator: #If it is the first operation to be performed, 
    text = display_state + " = " # the equals symbol is added to the expression to be inserted in the SuperField.
    superscript_text.set(text)   
    
  elif not operator: # If the operation performed reused the same right operand and the operator of the previous expression,
      text = input_field.get() + " " + ' '.join(re.findall(r'([-+*/]+)(\d+|\d+\.\d+)(=)', input_superfield.get().replace(' ',''))[0]) # only the left operand is changed. 
      superscript_text.set(text)  # And then it is inserted in the Super Field.
    
    
  clear_display(CE = True) # The text in the Input Field is deleted  
  
  input_field.insert(0,f"{result:,}") # and the result is inserted in the Input Field.


#----- objects StringVar for text in Input_field and Input_SuperField ----#
input_text = StringVar() 
superscript_text = StringVar()
#-----------------------#

#----- Frame for Input Field and SuperField ----#
input_frame = Frame( root, width= 312, height=100, bd=0,  bg= '#1a1a1a',)

input_frame.pack(side=TOP)                      
#-----------------------#

#----- Frame for the title ----#
text_field = Label(input_frame, text= 'Standart', background= "#1a1a1a",font=('Century Gothic', 10, 'normal' ),
                    width=10, height= 1 , bg= "#1b1b1b", bd=0,  foreground= "#ffffff")
text_field.pack(side= TOP, ipady=1, ipadx=1)
#-----------------------#

#----- Frame for the Text in the SuperField ----#
input_superfield = Entry(
                    input_frame, font=('Century Gothic', 15, 'normal' ), textvariable= superscript_text, 
                    width=100 , bg= "#1a1a1a", bd=0, justify= RIGHT,  foreground= "#ffffff",
                    )

input_superfield.pack(ipady=1)
#-----------------------#

#----- Frame for the text in the Input Text ----#
input_field = Entry(
                    input_frame, font=('Century Gothic', 40, 'normal' ), textvariable= input_text, 
                    width=100, bg= "#1a1a1a", bd=0, justify= RIGHT,  foreground= "#ffffff",
                    )

input_field.pack(ipady=1)
#-----------------------#

#----- Frame for Bottons ----#
btns_frame = Frame(root, width=400, height=375, bg= '#1b1b1b')

btns_frame.pack(side='bottom')
#-----------------------#

#----- Bottons -----#

# row 0 
percentage = Button( 
                    btns_frame, text= "\u0025", font=('Century Gothic', 15, 'bold' ),command = lambda: calculate("/100")
                    , 
                   fg= "#ffffff", width = 8, height=2, bd=0, bg= '#111111', cursor = 'hand2', 
                ).grid(row= 0, column=0 , padx=1, pady=1)


clear_entry = Button(
                     btns_frame, text= "CE", font=('Century Gothic', 15, 'bold' ), command = lambda: clear_display(CE = True),
                     fg= "#ffffff", width = 8, height=2, bd=0, bg= '#111111', cursor = 'hand2', 
                ).grid(row= 0, column=1 , padx=1, pady=1)

clear = Button(
              btns_frame, text= "C",font=('Century Gothic', 15, 'bold' ), command = lambda: clear_display(),
               fg= "#ffffff", width = 8, height=2, bd=0, bg= '#111111', cursor = 'hand2', 
                ).grid(row= 0, column=2 , padx=1, pady=1)

backspace = Button(
                    btns_frame, text="\u232B", font=('Century Gothic', 15, 'bold' ), command = lambda: clear_display(undo = True),
                    fg= "#ffffff", width = 8, height=2, bd=0, bg= '#111111', cursor = 'hand2', 
                ).grid(row= 0, column=3 , padx=1, pady=1)

# rows 1
reciprocal = Button(
              btns_frame, text= "\u00b9\u2044\U0001D465", font=('Century Gothic', 15, 'bold' ), command = lambda: calculate(operator="1/"),
               fg= "#ffffff", width = 8, height=2, bd=0, bg= '#111111', cursor = 'hand2', 
                ).grid(row= 1, column=0 , padx=1, pady=1)

power_two = Button(
              btns_frame, text= "\U0001D465\u00B2", font=('Century Gothic', 15, 'bold' ), command = lambda: calculate(operator="**2"),
               fg= "#ffffff", width = 8, height=2, bd=0, bg= '#111111', cursor = 'hand2', 
                ).grid(row= 1, column=1 , padx=1, pady=1)

square_two = Button(
              btns_frame, text= "\u00B2\u221A\U0001D465", font=('Century Gothic', 15, 'bold' ), command = lambda: calculate(operator="**0.5"),
               fg= "#ffffff", width = 8, height=2, bd=0, bg= '#111111', cursor = 'hand2', 
                ).grid(row= 1, column=2 , padx=1, pady=1)
divide = Button(
              btns_frame, text= "\u00F7",font=('Century Gothic', 15, 'bold' ), command = lambda: get_operation(operator="/"),
               fg= "#ffffff", width = 8, height=2, bd=0, bg= '#111111', cursor = 'hand2', 
                ).grid(row= 1, column=3 , padx=1, pady=1)

# rows 2 
seven = Button(
              btns_frame, text= "\u0037",font=('Century Gothic', 15, 'bold' ), command = lambda: get_numbers(7),
               fg= "#ffffff", width = 8, height=2, bd=0, bg= '#000000', cursor = 'hand2', 
                ).grid(row= 2, column=0 , padx=1, pady=1)

eight = Button(
              btns_frame, text= "\u0038",font=('Century Gothic', 15, 'bold' ), command = lambda : get_numbers(8),
               fg= "#ffffff", width = 8, height=2, bd=0, bg= '#000000', cursor = 'hand2', 
                ).grid(row= 2, column=1 , padx=1, pady=1)

nine = Button(
              btns_frame, text= "\u0039",font=('Century Gothic', 15, 'bold' ), command = lambda  : get_numbers(9),
               fg= "#ffffff", width = 8, height=2, bd=0, bg= '#000000', cursor = 'hand2', 
                ).grid(row= 2, column=2 , padx=1, pady=1)
multiply = Button(
              btns_frame, text= "\u00D7",font=('Century Gothic', 15, 'bold' ), command = lambda: get_operation(operator="*"),
               fg= "#ffffff", width = 8, height=2, bd=0, bg= '#111111', cursor = 'hand2', 
                ).grid(row= 2, column=3 , padx=1, pady=1)

# rows 3
four = Button(
              btns_frame, text= "\u0034",font=('Century Gothic', 15, 'bold' ), command = lambda  : get_numbers(4),
               fg= "#ffffff", width = 8, height=2, bd=0, bg= '#000000', cursor = 'hand2', 
                ).grid(row= 3, column=0 , padx=1, pady=1)

five = Button(
              btns_frame, text= "\u0035",font=('Century Gothic', 15, 'bold' ), command = lambda  : get_numbers(5),
               fg= "#ffffff", width = 8, height=2, bd=0, bg= '#000000', cursor = 'hand2', 
                ).grid(row= 3, column=1 , padx=1, pady=1)

six = Button(
              btns_frame, text= "\u0036",font=('Century Gothic', 15, 'bold' ), command = lambda  : get_numbers(6),
               fg= "#ffffff", width = 8, height=2, bd=0, bg= '#000000', cursor = 'hand2', 
                ).grid(row= 3, column=2 , padx=1, pady=1)
minus = Button(
              btns_frame, text= "\u2212",font=('Century Gothic', 15, 'bold' ), command = lambda : get_operation(operator="-"),
               fg= "#ffffff", width = 8, height=2, bd=0, bg= '#111111', cursor = 'hand2', 
                ).grid(row= 3, column=3 , padx=1, pady=1)

# rows 4
one = Button(
              btns_frame, text= "\u0031",font=('Century Gothic', 15, 'bold' ), command = lambda  : get_numbers(1),
               fg= "#ffffff", width = 8, height=2, bd=0, bg= '#000000', cursor = 'hand2', 
                ).grid(row= 4, column=0 , padx=1, pady=1)

two = Button(
              btns_frame, text= "\u0032",font=('Century Gothic', 15, 'bold' ), command = lambda  : get_numbers(2),
               fg= "#ffffff", width = 8, height=2, bd=0, bg= '#000000', cursor = 'hand2', 
                ).grid(row= 4, column=1 , padx=1, pady=1)

three = Button(
              btns_frame, text= "\u0033",font=('Century Gothic', 15, 'bold' ), command = lambda  : get_numbers(3),
               fg= "#ffffff", width = 8, height=2, bd=0, bg= '#000000', cursor = 'hand2', 
                ).grid(row= 4, column=2 , padx=1, pady=1)
plus = Button(
              btns_frame, text= "\u002B",font=('Century Gothic', 15, 'bold' ), command = lambda: get_operation(operator="+"),
               fg= "#ffffff", width = 8, height=2, bd=0, bg= '#111111', cursor = 'hand2', 
                ).grid(row= 4, column=3 , padx=1, pady=1)

# rows 5
inverse = Button(
              btns_frame, text= "\u207A\u2044\u208b",font=('Century Gothic', 15, 'bold' ), command = lambda : calculate(operator="*-1"),
               fg= "#ffffff", width = 8, height=2, bd=0, bg= '#000000', cursor = 'hand2', 
                ).grid(row= 5, column=0 , padx=1, pady=1)

zero = Button(
              btns_frame, text= "\u0030",font=('Century Gothic', 15, 'bold' ), command = lambda : get_numbers(0),
               fg= "#ffffff", width = 8, height=2, bd=0, bg= '#000000', cursor = 'hand2', 
                ).grid(row= 5, column=1 , padx=1, pady=1)

point = Button(
              btns_frame, text= "\u22C5",font=('Century Gothic', 15, 'bold' ), command = lambda : get_numbers("."),
               fg= "#ffffff", width = 8, height=2, bd=0, bg= '#000000', cursor = 'hand2', 
                ).grid(row= 5, column=2 , padx=1, pady=1)
equal = Button(
              btns_frame, text= "\u003D",font=('Century Gothic', 15, 'bold' ), command = lambda  :calculate(),
               fg= "#ffffff", width = 8, height=2, bd=0, bg= '#004185', cursor = 'hand2', 
                ).grid(row= 5, column=3 , padx=1, pady=1)

#----- LOOP ----#
# root.mainloop()
  
```
